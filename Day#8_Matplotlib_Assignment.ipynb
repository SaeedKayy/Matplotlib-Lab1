```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Q1: Create two charts using list and dataframe

```python
games = ['Halo', 'MW2', 'GoW', 'Stray']
rating = [8.2, 9.1, 7.5, 9.9]
rating2 = [8.5, 9.2, 7.7, 9.6]
rating3 = [8.7, 9.0, 7.8, 9.8]
df = pd.DataFrame(data={'Games': games, 'Ratings': rating, 'Ratings 2': rating2, 'Ratings 3': rating3 })
df
```

## Q2: Create pie chart with equal explode values.

```python
plt.pie(df['Ratings'], labels=df['Games'],explode = [0,0,0,0])
plt.show()
```

## Q3: Create a scatter chart with circle marker, alpha = 0.4, and size = 50.

```python
plt.scatter(df['Games'], df['Ratings'], alpha=0.4, s=50)
plt.show()
```

## Q4: Add the following to your chart:
- title placed in the left and has a red color
- x and y label with font size = 16

```python
plt.scatter(df['Games'], df['Ratings'], alpha=0.4, s=50)
plt.show()
plt.title('Games Rating', loc='left', color='red')
plt.xlabel('Games', fontsize=16)
plt.ylabel('Ratings', fontsize=16)
plt.show()
```

## Q5: Create three line charts with Legends at left and colors

```python
plt.plot(df['Games'], df['Ratings'], color='blue', label='rating 1')
plt.plot(df['Games'], df['Ratings 2'], color='red', label='rating 2')
plt.plot(df['Games'], df['Ratings 3'], color='green', label='rating 3')
plt.legend(loc='upper left')
plt.xlabel("Games")
plt.ylabel("RATING")
plt.show() 
```

## Q6: Create subplots with three charts and figsize=20,10

```python
fig, ax = plt.subplots(nrows=3, ncols=1, figsize=(20,10))

ax[0].plot(df['Games'], df['Ratings'], color='red', label='Rating 1')
ax[0].legend(loc='upper left')

ax[1].plot(df['Games'], df['Ratings 2'], color='blue', label='Rating 2')
ax[1].legend(loc='upper left')

ax[2].plot(df['Games'], df['Ratings 3'], color='green', label='Rating 3')
ax[2].legend(loc='upper left')

ax[0].set_xlabel("Games")
ax[0].set_ylabel("Rating")

plt.show()
```
